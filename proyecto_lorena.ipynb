### 3. Metodología Propuesta: Modelos de Deep Learning

Se explorarán dos enfoques principales de _Deep Learning_ para la clasificación de imágenes:

#### 3.1. Modelo Base con CNN desde Cero

El primer enfoque consiste en construir y entrenar una red neuronal convolucional (CNN) sencilla desde cero. Este modelo servirá como línea base para entender cómo una arquitectura básica de CNN se desempeña en la tarea de clasificación de imágenes de aves. Se utilizarán capas convolucionales, de _pooling_, y capas densas, junto con técnicas de regularización como _Dropout_ y aumento de datos (_Data Augmentation_).

#### 3.2. Modelo con Transfer Learning (ResNet50)

El segundo enfoque, y el principal para obtener un alto rendimiento, es el uso de _Transfer Learning_. Se utilizará un modelo pre-entrenado en un dataset grande como ImageNet, específicamente **ResNet50**, y se adaptará para la tarea de clasificación de aves.

**Pasos clave:**
1.  **Congelar Capas Base:** Las capas convolucionales del modelo ResNet50 se congelarán, aprovechando las características de bajo nivel aprendidas de ImageNet.
2.  **Añadir Capas Personalizadas:** Se añadirán capas densas (`GlobalAveragePooling2D`, `Dropout`, `Dense`) en la parte superior del modelo congelado para adaptarlo a las 200 clases específicas del dataset CUB-200-2011.
3.  **Entrenamiento de la Cabeza:** Inicialmente, solo se entrenarán las nuevas capas personalizadas.
4.  **Ajuste Fino (Fine-tuning):** (Opcional en esta fase inicial, pero considerado para futuras iteraciones) Una vez que la cabeza del modelo haya convergido, se descongelarán algunas de las capas superiores del modelo base y se entrenará el modelo completo con una tasa de aprendizaje muy baja para ajustar las características pre-entrenadas a nuestro dataset específico.

```python
import tensorflow as tf

IMG_SIZE = 224
BATCH_SIZE = 32
NUM_CLASSES = 200

image_dir = "/content/cub200_dataset/images/"

ds_train = tf.keras.utils.image_dataset_from_directory(
    image_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE
)

ds_validation = tf.keras.utils.image_dataset_from_directory(
    image_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE
)

print(f"Clases encontradas: {len(ds_train.class_names)}")
print(ds_train.class_names[:5])
```

```python
import numpy as np
import matplotlib.pyplot as plt

NUM_CLASSES = 200

# Obtener las etiquetas del dataset de entrenamiento
train_labels = np.concatenate([y.numpy() for x, y in ds_train], axis=0)

# Obtener las etiquetas del dataset de validación
val_labels = np.concatenate([y.numpy() for x, y in ds_validation], axis=0)

# Concatenar todas las etiquetas
all_labels = np.concatenate([train_labels, val_labels], axis=0)

# Contar la frecuencia de cada clase
class_counts = np.bincount(all_labels, minlength=NUM_CLASSES)

# Crear un gráfico de barras
plt.figure(figsize=(15, 6))
plt.bar(range(NUM_CLASSES), class_counts)
plt.xlabel('ID de Clase')
plt.ylabel('Número de Imágenes')
plt.title('Distribución de Clases en el Dataset CUB-200-2011')
plt.xticks(np.arange(0, NUM_CLASSES, 20), rotation=90) # Mostrar cada 20 clases para no saturar
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print(f"Número total de imágenes: {len(all_labels)}")
print(f"Mínimo de imágenes por clase: {np.min(class_counts[class_counts > 0])}")
print(f"Máximo de imágenes por clase: {np.max(class_counts)}")
print(f"Promedio de imágenes por clase: {np.mean(class_counts)}")
```

```python
import tensorflow as tf
from tensorflow.keras import layers

IMG_SIZE = 224
BATCH_SIZE = 32
NUM_CLASSES = 200

data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
], name="data_augmentation")

rescaling = layers.Rescaling(1./255, name="rescaling")


AUTOTUNE = tf.data.AUTOTUNE

ds_train = ds_train.cache().prefetch(buffer_size=AUTOTUNE)
ds_validation = ds_validation.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
model_base = tf.keras.Sequential([
    layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3)),

    rescaling,

    data_augmentation,

    layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),

    layers.Dense(NUM_CLASSES, activation='softmax')
])

model_base.summary()
```

```python
model_base.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Iniciando entrenamiento del modelo base...")

history_base_cnn = model_base.fit(
    ds_train,
    validation_data=ds_validation,
    epochs=15
)

print("Entrenamiento del modelo base finalizado.")
```

```python
import matplotlib.pyplot as plt

# --- 1. Resumen del historial de entrenamiento ---
acc = history_base_cnn.history['accuracy']
val_acc = history_base_cnn.history['val_accuracy']
loss = history_base_cnn.history['loss']
val_loss = history_base_cnn.history['val_loss']

epochs_range = range(len(acc))

# --- 2. Graficar Precisión ---
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Precisión de Entrenamiento')
plt.plot(epochs_range, val_acc, label='Precisión de Validación')
plt.legend(loc='lower right')
plt.title('Precisión de Entrenamiento vs. Validación')

# --- 3. Graficar Pérdida ---
plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Pérdida de Entrenamiento')
plt.plot(epochs_range, val_loss, label='Pérdida de Validación')
plt.legend(loc='upper right')
plt.title('Pérdida de Entrenamiento vs. Validación')
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers

IMG_SIZE = 224
NUM_CLASSES = 200

data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
], name="data_augmentation")

base_model_r50 = tf.keras.applications.ResNet50(
    include_top=False,
    weights='imagenet',
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)

base_model_r50.trainable = False

print("Capas base de ResNet50 cargadas y congeladas.")
```

```python
inputs = tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3), name="input_layer")

x = data_augmentation(inputs)

x = tf.keras.applications.resnet.preprocess_input(x)

x = base_model_r50(x, training=False)

x = layers.GlobalAveragePooling2D(name="global_avg_pool")(x)
x = layers.Dropout(0.5, name="regularization_dropout")(x)
outputs = layers.Dense(NUM_CLASSES, activation='softmax', name="output_layer")(x)
model_resnet = tf.keras.Model(inputs, outputs)

model_resnet.summary()
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

model_resnet.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Iniciando entrenamiento del modelo ResNet (solo la cabeza)...")

history_resnet = model_resnet.fit(
    ds_train,
    validation_data=ds_validation,
    epochs=20
)

print("Entrenamiento del modelo ResNet finalizado.")
```

```python
acc = history_resnet.history['accuracy']
val_acc = history_resnet.history['val_accuracy']
loss = history_resnet.history['loss']
val_loss = history_resnet.history['val_loss']

epochs_range = range(len(acc))

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Precisión de Entrenamiento (ResNet)')
plt.plot(epochs_range, val_acc, label='Precisión de Validación (ResNet)')
plt.legend(loc='lower right')
plt.title('Precisión - Modelo ResNet')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Pérdida de Entrenamiento (ResNet)')
plt.plot(epochs_range, val_loss, label='Pérdida de Validación (ResNet)')
plt.legend(loc='upper right')
plt.title('Pérdida - Modelo ResNet')
plt.show()
```

## Overfitting


```python
from tensorflow.keras.callbacks import EarlyStopping

# Definir el callback de EarlyStopping
# Monitoreamos la 'val_loss' (pérdida de validación)
# Si la 'val_loss' no mejora durante 5 épocas consecutivas, el entrenamiento se detendrá.
# 'restore_best_weights=True' asegura que se carguen los pesos del mejor modelo (antes de que comenzara a sobreajustarse).
early_stopping_callback = EarlyStopping(
    monitor='val_loss',
    patience=5, # Número de épocas sin mejora para detener el entrenamiento
    restore_best_weights=True # Restaurar los pesos del modelo de la mejor época
)

print("Callback de Early Stopping definido.")
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau

# Obtener el modelo base ResNet50 dentro de model_resnet
resnet_base_model = model_resnet.get_layer('resnet50')

# Congelar las primeras capas del modelo base (hasta el índice 140)
for layer in resnet_base_model.layers[:140]:
    layer.trainable = False

# Descongelar las capas restantes del modelo base (desde el índice 140 en adelante)
for layer in resnet_base_model.layers[140:]:
    layer.trainable = True

print("Capas del modelo base ResNet50 configuradas para ajuste fino estratégico.")

# Definir el callback ReduceLROnPlateau
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss', # Monitorear la pérdida de validación
    factor=0.2,         # Reducir la tasa de aprendizaje a un 20% de su valor actual
    patience=3,         # Esperar 3 épocas sin mejora antes de reducir
    min_lr=1e-8,        # Tasa de aprendizaje mínima
    verbose=1           # Mostrar mensajes cuando la tasa de aprendizaje se reduce
)

print("Callback ReduceLROnPlateau definido.")

# Recompilar el modelo con una tasa de aprendizaje muy baja para el ajuste fino
optimizer_finetune_strategic = tf.keras.optimizers.Adam(learning_rate=1e-7)

model_resnet.compile(
    optimizer=optimizer_finetune_strategic,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Modelo ResNet recompilado para ajuste fino estratégico con una tasa de aprendizaje muy baja.")

print("Iniciando ajuste fino del modelo ResNet con ReduceLROnPlateau...")

history_resnet_finetune_v2 = model_resnet.fit(
    ds_train,
    validation_data=ds_validation,
    epochs=1, # Entrenar por 20 épocas adicionales
    callbacks=[early_stopping_callback, reduce_lr] # Usar EarlyStopping y ReduceLROnPlateau
)

print("Ajuste fino del modelo ResNet finalizado.")
```

3.  **Entrenamiento de la Cabeza:** Inicialmente, solo se entrenarán las nuevas capas personalizadas.
4.  **Ajuste Fino (Opcional):** Una vez que la cabeza haya convergido, se puede realizar un ajuste fino de las capas superiores del modelo base con una tasa de aprendizaje muy baja.

```python
import tensorflow as tf
from tensorflow.keras import layers

# Definir los parámetros, si no están ya definidos
# IMG_SIZE = 224
# NUM_CLASSES = 200
# BATCH_SIZE = 32

# La capa de aumento de datos y reescalado ya deberían estar definidas
# en una celda anterior (yp_Gv3lgryjJ)
# Si no lo están, puedes descomentar y ejecutar:
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
], name="data_augmentation")

# # No necesitamos rescaling aquí si EfficientNet tiene su propio preprocess_input
# # pero si quieres uniformidad o el modelo lo requiere, puedes usarlo antes del preprocess_input
# # rescaling = layers.Rescaling(1./255, name="rescaling")

# Cargar el modelo base EfficientNetB0 pre-entrenado
base_model_efficientnet = tf.keras.applications.EfficientNetB0(
    include_top=False, # No incluir las capas de clasificación superiores
    weights='imagenet', # Usar pesos pre-entrenados en ImageNet
    input_shape=(IMG_SIZE, IMG_SIZE, 3) # Definir la forma de entrada
)

# Congelar las capas del modelo base para Transfer Learning
base_model_efficientnet.trainable = False

print("Capas base de EfficientNetB0 cargadas y congeladas.")
```

```python
inputs_efficientnet = tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3), name="input_layer_efficientnet")

# Aplicar aumento de datos
x = data_augmentation(inputs_efficientnet)

# Preprocesamiento específico para EfficientNet
# EfficientNet ya maneja su propio escalado internamente o se espera entrada en [0, 255]
# Sin embargo, Keras Applications modelos a menudo tienen un preprocess_input
# Para EfficientNet, si los pesos son 'imagenet', se espera entrada en [0, 255] o [-1, 1] dependiendo de la versión.
# La capa Rescaling 1./255 es generalmente buena para la mayoría de los modelos.

#x = rescaling(x) # Aplicamos la capa de reescalado ya definida

# Pasar la entrada preprocesada a través del modelo base congelado
x = base_model_efficientnet(x, training=False)

# Añadir capas personalizadas para la clasificación de aves
x = layers.GlobalAveragePooling2D(name="global_avg_pool_efficientnet")(x)

x = layers.Dropout(0.3, name="regularization_dropout_efficientnet")(x) # Ajustar dropout si es necesario
outputs_efficientnet = layers.Dense(NUM_CLASSES, activation='softmax', name="output_layer_efficientnet")(x)

# Construir el modelo completo
model_efficientnet = tf.keras.Model(inputs_efficientnet, outputs_efficientnet)

model_efficientnet.summary()
```

```python
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

optimizer_efficientnet = tf.keras.optimizers.Adam(learning_rate=0.001)

model_efficientnet.compile(
    optimizer=optimizer_efficientnet,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Definir el callback de EarlyStopping
early_stopping_efficientnet_head = EarlyStopping(
    monitor='val_loss',
    patience=7, # Número de épocas sin mejora para detener el entrenamiento
    restore_best_weights=True # Restaurar los pesos del modelo de la mejor época
)

# Definir el callback ReduceLROnPlateau
reduce_lr_efficientnet_head = ReduceLROnPlateau(
    monitor='val_loss', # Monitorear la pérdida de validación
    factor=0.2,         # Reducir la tasa de aprendizaje a un 20% de su valor actual
    patience=3,         # Esperar 3 épocas sin mejora antes de reducir
    min_lr=1e-8,        # Tasa de aprendizaje mínima
    verbose=1           # Mostrar mensajes cuando la tasa de aprendizaje se reduce
)

print("Iniciando entrenamiento del modelo EfficientNet (solo la cabeza) con EarlyStopping y ReduceLROnPlateau...")

history_efficientnet = model_efficientnet.fit(
    ds_train,
    validation_data=ds_validation,
    epochs=30, # Puedes ajustar el número de épocas, EarlyStopping lo detendrá
    callbacks=[early_stopping_efficientnet_head, reduce_lr_efficientnet_head]
)

print("Entrenamiento del modelo EfficientNet finalizado.")
```
